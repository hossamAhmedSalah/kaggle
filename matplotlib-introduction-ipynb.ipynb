````python
# another parameter of figure 
# DPI (dots per inch) a measure of the resolution of a digital image
# the default DPI in matplotlib is 100 
# if you were to save a figure as an image fileüñºÔ∏è, such as a PNG or JPEG, it would have 100 pixels per inch
# how to change the dpi of a figure ‚ùì 
# 1Ô∏è‚É£ you can pass a parameter 
fig2 = plt.figure(figsize=(5, 4), dpi=150)
plt.bar(lables, values)
````

#### Why the size of the chart change ? although we have the same ```figsize =(5, 7)``` like in fig ‚ùì 
The DPI (dots per inch) of an image determines how many pixels will be displayed per inch when the image is printed or displayed on a screen. When you change the DPI of an image, you are essentially changing the physical size of the image as well.
In the context of Matplotlib, changing the DPI of a figure will also change the physical size of the figure. 

Therefore, when you change the DPI of an image or a Matplotlib figure, you are essentially changing the trade-off between image quality and image size. A higher DPI will result in a higher quality image, but it will also result in a larger physical size. Conversely, a lower DPI will result in a lower quality image, but it will also result in a smaller physical size.

### How to style the line ?
```linestyle```: A string specifying the style of the line. Valid values are '-' (solid line), '--' (dashed line), ':' (dotted line), '-.' (dash-dot line), or 'None' (no line). The default value is '-'.

```linewidth```: A scalar specifying the width of the line in points. The default value is 1.0.

```color```: A string specifying the color of the line. Valid values are standard color names like 'red', 'blue', and 'green', as well as abbreviations like 'r', 'b', and 'g'. The default value is 'b' (blue).

```marker```: A string specifying the shape of the marker to be used at the data points. Valid values are '.', 'o', 'x', '+', and many others. The default value is 'None' (no marker).

```markersize```: A scalar specifying the size of the markers in points. The default value is 6.0.

```markeredgecolor```: A string specifying the color of the marker edges. The default value is 'None' (no edge color).

```markeredgewidth```: A scalar specifying the width of the marker edges in points. The default value is 1.0.

```markerfacecolor```: A string specifying the color of the marker faces. The default value is 'None' (no face color).

### Scatter chart

#### ```plt.scatter() parameters``` 
x: A 1D array or list of values representing the x-coordinates of the data points.

y: A 1D array or list of values representing the y-coordinates of the data points.

s: A scalar or 1D array or list of values representing the size of the markers. If s is a scalar, all markers will have the same size. If s is an array or list, the size of each marker will be determined by the corresponding value in s.

c: A scalar or 1D array or list of values representing the color of the markers. If c is a scalar, all markers will have the same color. If c is an array or list, the color of each marker will be determined by the corresponding value in c. c can also be a string or RGB tuple representing a single color that will be used for all markers.

marker: A string or a Path object representing the shape of the markers. Some common marker styles include 'o' (circle), 's' (square), '+' (plus), and 'x' (cross).

alpha: A scalar between 0 and 1 representing the opacity of the markers. A value of 1 indicates that the markers are fully opaque, while a value of 0 indicates that the markers are fully transparent.

linewidths: A scalar or 1D array or list of values representing the width of the marker edges.

edgecolors: A string or 1D array or list of colors representing the color of the marker edges.

cmap: A string or a Colormap object representing the color map used to map scalar values to colors. Some common color maps include 'viridis', 
'cool', and 'jet'. 
more about cmaps 'color maps'
ü™∂ https://matplotlib.org/stable/tutorials/colors/colormaps.html#sphx-glr-tutorials-colors-colormaps-py

````python
fig = plt.figure(figsize=(8, 5)) # initialize figure

ax = [None for _ in range(6)] # list to save many ax for setting parameter in each

ax[0] = plt.subplot2grid((3,4), (0,0), colspan=4)
ax[1] = plt.subplot2grid((3,4), (1,0), colspan=1)
ax[2] = plt.subplot2grid((3,4), (1,1), colspan=1)
ax[3] = plt.subplot2grid((3,4), (1,2), colspan=1)
ax[4] = plt.subplot2grid((3,4), (1,3), colspan=1,rowspan=2)
# ax[5] = plt.subplot2grid((3,4), (2,0), colspan=3)
ax[5] = plt.subplot2grid((3,4), (2,1))


for ix in range(6): 
    ax[ix].set_title('ax[{}]'.format(ix)) # make ax title for distinguish:)
    ax[ix].set_xticks([]) # to remove x ticks
    ax[ix].set_yticks([]) # to remove y ticks
    
fig.tight_layout()
plt.show()
````

## gridspec  <span style="color:green">(EXTRA)</span>
more like slicing 

````python
fig = plt.figure(figsize=(8, 5))

gs = fig.add_gridspec(3, 3) # make 3 by 3 grid (row, col)

ax = [None for _ in range(5)]
# row 0 , all cols
ax[0] = fig.add_subplot(gs[0, :]) 
ax[0].set_title('gs[0, :]')

# row 1, till the last col but is not with us
ax[1] = fig.add_subplot(gs[1, :-1])
ax[1].set_title('gs[1, :-1]')

ax[2] = fig.add_subplot(gs[1:, -1])
ax[2].set_title('gs[1:, -1]')

ax[3] = fig.add_subplot(gs[-1, 0])
ax[3].set_title('gs[-1, 0]')

ax[4] = fig.add_subplot(gs[-1, -2])
ax[4].set_title('gs[-1, -2]')
# removing ticks 
for ix in range(5):
    ax[ix].set_xticks([])
    ax[ix].set_yticks([])
# setting a suitable speces between axes to not overlap (labels or ticks)
plt.tight_layout()
# show
plt.show()
````
